```python
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
raw_data = datasets.load_wine()
print(raw_data.DESCR)
```

```python
for key,value in raw_data.items():
    print(key,'\n',value,'\n')
```

```python
data = pd.DataFrame(data=raw_data.data, columns=raw_data.feature_names)
data
```

```python
data['type'] = raw_data['target']
data.head()
```

```python
sns.pairplot(data, hue='type')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(data.iloc[:,:-1],data['type'],
                                                                  test_size=0.3, random_state = 0)
print(len(X_train),' samples in training data\n', len(X_test),' samples in test data\n', )
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import validation_curve
import matplotlib.pyplot as plt

param = {'max_depth':list(range(1,6))}
grid = GridSearchCV(DecisionTreeClassifier(random_state=0),param_grid=param,cv = 5,refit=True)
grid.fit(X_train,Y_train)

print('Best max_depth is {}'.format(grid.best_params_['max_depth']))
plt.title('max_depth vs mean_test_accuracy')
plt.xlabel('max_depth')
plt.ylabel('mean_test_accuracy')
plt.plot(param['max_depth'], grid.cv_results_['mean_test_score'], '-o')
plt.grid()
for xy in zip(param['max_depth'], grid.cv_results_['mean_test_score']):
    plt.annotate('{:.5}'.format(xy[1]), textcoords='data', xy = xy)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix

dct = DecisionTreeClassifier(max_depth = 3, random_state=0) #
dct.fit(X_train, Y_train)
Y_pred = dct.predict(X_test)
cf_dct = confusion_matrix(Y_test,Y_pred)
acc = np.mean(np.array(Y_test) == Y_pred)
print('accuracy = {}'.format(acc))
print(cf_dct)
```

```python
from sklearn.tree.export import export_graphviz
export_graphviz(dct)

from sklearn.externals.six import StringIO  
import pydot
dot_data = StringIO() 
export_graphviz(dct, out_file=dot_data,filled = True, rounded = True, impurity = False,
               feature_names = raw_data['feature_names'], class_names = raw_data['target_names'])
graph = pydot.graph_from_dot_data(dot_data.getvalue()) 
graph[0].write_png('wine.png')
```
